### Data Cleaning: Make sure discrete and continous columns are converted to numerical datatype

**There are 3 main things that we need to do with regards to the datatypes:**
1. Convert the continuous and discrete variables to numeric data types 
2. Convert the nominal data columns to numeric data types by using get_dummies method
3. Convert the ordinal columns to numeric data types by assigning numerical values to them 

```python
#converting continous variables to numeric data type
num_columns = ['lot_frontage','lot_area','mas_vnr_area','bsmtfin_sf_1','bsmtfin_sf_2','bsmt_unf_sf','total_bsmt_sf',
               '1st_flr_sf','2nd_flr_sf','low_qual_fin_sf','gr_liv_area','garage_area','wood_deck_sf','open_porch_sf',
               'enclosed_porch','3ssn_porch','screen_porch','pool_area','misc_val','saleprice']
train[num_columns] = train[num_columns].apply(pd.to_numeric)
```

```python
#fixing discrete numerical variables
num2_columns=['bsmt_full_bath','bsmt_half_bath','garage_cars','garage_yr_blt']
train[num2_columns] = train[num2_columns].apply(pd.to_numeric)
```

At this point, we can also convert the ordinal columns to numerical values 

### Convert ordinal columns to numerical order

```python
ordinal_columns = ['lot_shape','utilities','land_slope','exter_qual',
                   'exter_cond','bsmt_qual','bsmt_cond','bsmt_exposure','bsmtfin_type_1',
                   'bsmtfin_type_2','heating_qc','electrical','kitchen_qual','functional',
                   'fireplace_qu','garage_finish','garage_qual','garage_cond','paved_drive','pool_qc','fence',
                   'overall_qual','overall_cond']
#overall_qual and overall_cond are already in numerical format
```

```python
len(ordinal_columns)
```

```python
#grouping those with the same measurement format 
ex_gd_ta_fa_po_na=['exter_qual','exter_cond','bsmt_qual','bsmt_cond','heating_qc','kitchen_qual','fireplace_qu',
                   'garage_qual','garage_cond','pool_qc']


#have to convert these columns manually since they have different categories
#QLC_rating = ['bsmtfin_type_1','bstmtfin_type_2',]
#bsmt exposure is on its own
#lot shape is on its own
#utilities is on its own
#land_slope
#electrical
#functional
#garage finish
#paved drive
#fence
```

```python
#to make sure they all have the same type of values
np.unique(train[ex_gd_ta_fa_po_na])
```

```python
def convert_columns(columns):
    '''to convert columns with categories Ex,Gd,TA,Fa,Po,NA to 5,4,3,2,1,0 respectively'''
    for column in columns:
        train[column] = train[column].map({'Ex':5,'Gd':4,'TA':3, 'Fa':2,'Po':1,'NA':0})
    return train

```

```python
convert_columns(ex_gd_ta_fa_po_na)
```

```python
#converting bsmt exposure, lot shape, utilities, landslope, 
#electrical, functional, garage finish, paved drive, fence, bsmtfin type 1 and bsmt fin type2
#to numerical column

train['bsmt_exposure'] = train['bsmt_exposure'].map({'Gd':4,'Av':3,'Mn':2, 'No':1,'NA':0})
train['lot_shape'] = train['lot_shape'].map({'Reg':4,'IR1':3,'IR2':2, 'IR3':1})
train['utilities'] = train['utilities'].map({'AllPub':4,'NoSewr':3,'NoSeWa':2, 'ELO':1})
train['land_slope'] = train['land_slope'].map({'Gtl':3,'Mod':2, 'Sev':1})
train['electrical'] = train['electrical'].map({'SBrkr':5,'FuseA':4, 'FuseF':3, 'FuseP':2, 'Mix':1})
train['functional'] = train['functional'].map({'Typ':8,'Min1':7, 'Min2':6, 'Mod':5,'Maj1':4, 'Maj2':3,
                                                       'Sev':2, 'Sal':1})
train['garage_finish'] = train['garage_finish'].map({'Fin':3,'RFn':2, 'Unf':1,'NA':0}) 
train['paved_drive'] = train['paved_drive'].map({'Y':3,'P':2, 'N':1}) 
train['fence'] = train['fence'].map({'GdPrv':4,'MnPrv':3,'GdWo':2, 'MnWw':1,'NA':0})
train['bsmtfin_type_1'] = train['bsmtfin_type_1'].map({'GLQ':6, 'ALQ':5,'BLQ':4, 'Rec':3,
                                                       'LwQ':2, 'Unf':1,'NA':0})
train['bsmtfin_type_2'] = train['bsmtfin_type_2'].map({'GLQ':6, 'ALQ':5,'BLQ':4, 'Rec':3,
                                                       'LwQ':2, 'Unf':1,'NA':0})
```

Up to this point, we have converted all **the continuous, discrete and ordinal variables to numeric datatypes.** Before we  move on to the nominal variables, I would like to do an **initial feature selection of the numerical datatypes** that we have to avoid cluttering the dataframe too much.

```python
def boxplot_discrete (list_of_columns,rows,columns,list_of_xlabels):
    '''create boxplit when a list of columns is passed as an argument, need to specify no of rows and columns of subplot'''
    fig, axes = plt.subplots(nrows=rows, ncols=columns,figsize=(12,16)) 
    fig.tight_layout(pad=5.0) #so that each subplot is well-spaced
    c=0
    for i,row in enumerate(axes):
        for j,col in enumerate(row):
            boxplot = sns.boxplot(data=train,x=list_of_columns[c],y="saleprice",ax=axes[i][j])
            boxplot.set_xlabel(list_of_xlabels[c],fontsize=11)
            boxplot.set_ylabel("Sale Price",fontsize=11)
            boxplot.set_title("Distributions summary",fontsize=15)
            c += 1
```

- as for fireplace, we can see that **the effect of the no of fireplace plateau after it reaches 1.** We may consider switching it to categorical variable (whether there is a fireplace or not as it seems like the number of fireplaces do not matter as much). We will confirm this in the statistics table below

```python
#created another function to draw boxplot for nominal variables
#did not reuse that function so that it is easier for me to change some parameters without scrolling up and down
def boxplot_nominal (list_of_columns,rows,columns,list_of_xlabels):
    ''' create boxplot when a list of columns is passed as an argument, need to specify no of rows and columns'''
    fig, axes = plt.subplots(nrows=rows, ncols=columns,figsize=(12,10)) 
    fig.tight_layout(pad=5.0) #so that each subplot is well-spaced
    c=0
    for i,row in enumerate(axes):
        for j,col in enumerate(row):
            boxplot = sns.boxplot(data=train,x=list_of_columns[c],y="saleprice",ax=axes[i][j])
            
            #setting the appropriate labels
            boxplot.set_xlabel(list_of_xlabels[c],fontsize=12)
            boxplot.set_ylabel("Sale Price",fontsize=11)
            boxplot.set_title("Distributions summary",fontsize=15)
            
            c += 1
            if c>=len(list_of_columns):
                break
```

## Final Features Selection from RFE

Since the lr model is doing slightly better than the rest, we will move on to use it to fit our model.

```python
X_lr_test.shape
```

```python
#finding the 48 features that have been selected by RFECV using lr model 
final_features = X_train.columns[rfecv_lr.support_]
```

```python
#pairing them with the coefficient
features_dict = dict(zip(final_features, rfecv_lr.estimator_.coef_))
```

```python
#finding the top 30 strongest coefficients from those 48 features

abs_features = {key: abs(value) for key,value in features_dict.items() }

from collections import Counter
k= Counter(abs_features)
high= k.most_common(30)

top_30=[]
for i in high:
    top_30.append(i[0])

top_30
```

```python
#Our new X_train and X_test should have 30 features only
X_train_new = ss.fit_transform(X_train[top_30])
X_test_new = ss.transform(X_test[top_30])
```
